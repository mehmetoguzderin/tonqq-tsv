```python
df = pd.read_csv("https://raw.githubusercontent.com/mehmetoguzderin/tonqq-tsv/main/tonqq.tsv", sep="\t", header=0, index_col=0)
df.index = df.index + 1
df['text'] = df['text'].apply(lambda i: _normal(i))
df['stela'] = df['stela'].apply(lambda i: i + 1)
df['face'] = df['face'].apply(lambda i: i + 1)
df['line'] = df['line'].apply(lambda i: i + 1)
df = df.rename_axis('id')
df
```

```python
df.to_csv("tonqq_text.csv")
```
